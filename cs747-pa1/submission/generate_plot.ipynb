```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
observation_file= pd.read_csv('output.txt',header=None)
len(observation_file)
```

```python
observation_file.head()
```

```python
observation_instance_1 = observation_file[observation_file[0]=='../instances/i-1.txt']
observation_instance_2 = observation_file[observation_file[0]=='../instances/i-2.txt']
observation_instance_3 = observation_file[observation_file[0]=='../instances/i-3.txt']
```

```python
round_robin_1 = observation_instance_1[observation_instance_1[1]==' round-robin']
epsilon_1 = observation_instance_1[observation_instance_1[1]==' epsilon-greedy']
ucb_1 = observation_instance_1[observation_instance_1[1]==' ucb']
kl_ucb_1 = observation_instance_1[observation_instance_1[1]==' kl-ucb']
thomson_1 = observation_instance_1[observation_instance_1[1]==' thompson-sampling']

round_robin_2 = observation_instance_2[observation_instance_2[1]==' round-robin']
epsilon_2 = observation_instance_2[observation_instance_2[1]==' epsilon-greedy']
ucb_2 = observation_instance_2[observation_instance_2[1]==' ucb']
kl_ucb_2 = observation_instance_2[observation_instance_2[1]==' kl-ucb']
thomson_2 = observation_instance_2[observation_instance_2[1]==' thompson-sampling']

round_robin_3 = observation_instance_3[observation_instance_3[1]==' round-robin']
epsilon_3 = observation_instance_3[observation_instance_3[1]==' epsilon-greedy']
ucb_3 = observation_instance_3[observation_instance_3[1]==' ucb']
kl_ucb_3 = observation_instance_3[observation_instance_3[1]==' kl-ucb']
thomson_3 = observation_instance_3[observation_instance_3[1]==' thompson-sampling']

epsilon_1_1 = epsilon_1[epsilon_1[3]==0.002]
epsilon_2_1 = epsilon_1[epsilon_1[3]==0.02]
epsilon_3_1 = epsilon_1[epsilon_1[3]==0.2]

epsilon_1_2 = epsilon_2[epsilon_2[3]==0.002]
epsilon_2_2 = epsilon_2[epsilon_2[3]==0.02]
epsilon_3_2 = epsilon_2[epsilon_2[3]==0.2]

epsilon_1_3 = epsilon_3[epsilon_3[3]==0.002]
epsilon_2_3 = epsilon_3[epsilon_3[3]==0.02]
epsilon_3_3 = epsilon_3[epsilon_3[3]==0.2]
```

```python
horizon = [50, 200, 800, 3200, 12800, 51200, 204800]
algorithms_1 = [round_robin_1, epsilon_1_1, epsilon_2_1, epsilon_3_1, ucb_1, kl_ucb_1, thomson_1]
algorithms_2 = [round_robin_2, epsilon_1_2, epsilon_2_2, epsilon_3_2, ucb_2, kl_ucb_2, thomson_2]
algorithms_3 = [round_robin_3, epsilon_1_3, epsilon_2_3, epsilon_3_3, ucb_3, kl_ucb_3, thomson_3]
```

```python
instance_1_dict = {}
for i in range(len(algorithms_1)):
    instance_1_dict[i] = []
    
instance_2_dict = {}
for i in range(len(algorithms_2)):
    instance_2_dict[i] = []

instance_3_dict = {}
for i in range(len(algorithms_3)):
    instance_3_dict[i] = []
    
for i,algorithms in enumerate(algorithms_1):
    for j in horizon:
        instance_1_dict[i].append(np.mean(algorithms[algorithms[4]==j][5]))
        
for i,algorithms in enumerate(algorithms_2):
    for j in horizon:
        instance_2_dict[i].append(np.mean(algorithms[algorithms[4]==j][5]))

for i,algorithms in enumerate(algorithms_3):
    for j in horizon:
        instance_3_dict[i].append(np.mean(algorithms[algorithms[4]==j][5]))
```

```python
x_axis = []
for i in horizon:
    x_axis.append(i)
```

```python
plt.figure(figsize=(13,8))
plt.loglog(x_axis,instance_1_dict[0], "r-*", label = 'round-robin', markersize=10)
plt.loglog(x_axis,instance_1_dict[3], "y-*", label = 'epsilon-greedy, e=0.2',markersize=10)
plt.loglog(x_axis,instance_1_dict[2], "b-*", label = 'epsilon-greedy, e=0.02',markersize=10)
plt.loglog(x_axis,instance_1_dict[1], "g-*", label = 'epsilon-greedy, e=0.002',markersize=10)
plt.loglog(x_axis,instance_1_dict[4], "k-*", label = 'ucb',markersize=10)
plt.loglog(x_axis,instance_1_dict[5], "m-*",label = 'kl-ucb',markersize=10)
plt.loglog(x_axis,instance_1_dict[6], "c-*", label = 'thompson-sampling',markersize=10)

#plt.tight_layout()
plt.xlabel('Horizon(log-scale)')
plt.ylabel('regret')
plt.legend()
plt.legend(loc='best')
plt.savefig('instance_1.png')
plt.show()
```
