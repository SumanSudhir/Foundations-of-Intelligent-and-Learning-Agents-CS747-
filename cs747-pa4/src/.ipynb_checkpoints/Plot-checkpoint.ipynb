```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_a = pd.read_csv("output_a.txt", header=None)
df_b = pd.read_csv("output_b.txt", header=None)
df_c = pd.read_csv("output_c.txt", header=None)
```

```python
plt.figure(figsize=(13,8))
plt.plot(df_a[3],df_a[1], "r", label = 'windyGridWorld_part_a')#, markersize=10)
# plt.plot(mean_time_b,episode_b, "g--", label = 'windyGridWorld2')#, markersize=10)
plt.xlabel('time')
plt.ylabel('episode')
plt.legend()
plt.legend(loc='best')
plt.savefig('windyGridWorld_part_a')
```

```python
plt.figure(figsize=(13,8))
plt.plot(df_b[3],df_b[1], "b", label = 'windyGridWorld_part_b')#, markersize=10)
# plt.plot(mean_time_b,episode_b, "g--", label = 'windyGridWorld2')#, markersize=10)
plt.xlabel('time')
plt.ylabel('episode')
plt.legend()
plt.legend(loc='best')
plt.savefig('windyGridWorld_part_b')
```

```python
plt.figure(figsize=(13,8))
plt.plot(df_c[3],df_c[1], "g", label = 'windyGridWorld_part_c')#, markersize=10)
# plt.plot(mean_time_b,episode_b, "g--", label = 'windyGridWorld2')#, markersize=10)
plt.xlabel('time')
plt.ylabel('episode')
plt.legend()
plt.legend(loc='best')
plt.savefig('windyGridWorld_part_c')
```
